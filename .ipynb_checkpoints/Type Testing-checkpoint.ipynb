### Test Play-by-Play

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2018_play-by-play.html'
browser.get(url)
```

```python
year_df = pd.read_html(url)[0]
year_df
```

```python
year_df.columns.levels[0]
```

```python
year_df.columns = year_df.columns.droplevel()
```

```python
year_df
```

```python
# Change columns with multiple names (Shoot, Off.)
cols = []
scount, ocount = 1, 1
for column in year_df.columns:
    if column == 'Shoot':
        cols.append(f'Shoot_{scount}')
        scount+=1
        continue
    elif column == 'Off.':
        cols.append(f'Off_{ocount}')
        ocount+=1
        continue
    cols.append(column)
year_df.columns = cols
```

```python
year_df = year_df.rename(columns={
    'OnCourt':'PMP100_OnCourt',
    'On-Off':'PMP100_On_Off',
    'BadPass': 'BadPassTO',
    'LostBall': 'LostBallTO',
    'Shoot_1': 'FoulC_Shoot',
    'Shoot_2': 'FoulD_Shoot',
    'Off_1': 'FoulC_Off',
    'Off_2': 'FoulD_Off',
})
year_df
```

```python
text_keys = ['Player','Pos','Tm']
keys_list = year_df.keys()
num_keys = []
for key in keys_list:
    if key not in text_keys:
        num_keys.append(key)
year_df = year_df.loc[year_df["Rk"] != "Rk"]
```

### Need to transfrom Position % to number before converting column data type

```python
perc_list = ["PG%","SG%","SF%","PF%","C%"]
for col in perc_list:
    year_df[col] = pd.to_numeric(year_df[col].replace('\%','',regex=True))/100
    #year_df.loc[:, col] = col_fixer
# pd.to_numeric(year_df["PG%"].replace('\%','',regex=True))/100
# pd.to_numeric(year_df["SG%"].replace('\%','',regex=True))/100
# pd.to_numeric(year_df["SF%"].replace('\%','',regex=True))/100
# pd.to_numeric(year_df["PF%"].replace('\%','',regex=True))/100
# pd.to_numeric(year_df["C%"].replace('\%','',regex=True))/100
```
