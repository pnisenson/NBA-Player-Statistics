### Test Play-by-Play

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2018_play-by-play.html'
browser.get(url)
```

```python
year_df = pd.read_html(url)[0]
year_df
```

```python
year_df.columns.levels[0]
```

```python
year_df.columns = year_df.columns.droplevel()
```

```python
year_df
```

```python
# Change columns with multiple names (Shoot, Off.)
cols = []
scount, ocount = 1, 1
for column in year_df.columns:
    if column == 'Shoot':
        cols.append(f'Shoot_{scount}')
        scount+=1
        continue
    elif column == 'Off.':
        cols.append(f'Off_{ocount}')
        ocount+=1
        continue
    cols.append(column)
year_df.columns = cols
```

```python
year_df = year_df.rename(columns={
    'OnCourt':'PMP100_OnCourt',
    'On-Off':'PMP100_On_Off',
    'BadPass': 'BadPassTO',
    'LostBall': 'LostBallTO',
    'Shoot_1': 'FoulC_Shoot',
    'Shoot_2': 'FoulD_Shoot',
    'Off_1': 'FoulC_Off',
    'Off_2': 'FoulD_Off',
})
year_df
```

```python
year_df.keys()
```

```python
year_df = year_df.loc[year_df["Rk"] != "Rk"]
text_keys = ['Player','Pos','Tm']
keys_list = year_df.keys()
num_keys = []
for key in keys_list:
    if key not in text_keys:
        num_keys.append(key)
for key in num_keys:
    try:
        year_df[key] = pd.to_numeric(year_df[key])
    except:
        year_df[key] = pd.to_numeric(year_df[key].replace('\%','',regex=True))/100

```

```python
year_df.dtypes
```
